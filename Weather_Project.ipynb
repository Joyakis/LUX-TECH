```python
import pandas as pd
data = pd.read_csv("1. Weather Data.csv")
```

**OBJECTIVES**



*   Find all records where the weather was exactly clear.

*   Find the number of times the wind speed was exactly 4 km/hr

*   Check if there are any NULL values present in the dataset.
*   Rename the column "Weather" to "Weather_Condition."


*   What is the mean visibility of the dataset?


*   Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km


*   What is the mean value of each column for each weather condition?
*   Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.


*   Find the number of weather conditions that include snow.









```python
data.head()
```

```python
data['Weather'].value_counts()
```

```python
data.info()
```

```python
# Filter rows where the 'weather' column is 'clear'
clear_days = data[data['Weather'] == 'Clear']

# Display the rows where the weather is 'clear'
#print(clear_days)

# Display the count of rows where the weather is 'clear'
print(f"Count of clear days: {clear_days.shape[0]}")
```

```python
wind_speed_4_km = data[data['Wind Speed_km/h'] == 4]
print(f"Count of wind speed 4 km/hr: {wind_speed_4_km.shape[0]}")
```

```python
#checking for null values
data.isna().sum()
```

```python
#Renaming a column
data.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
```

```python
data.info()
```

```python
#finding the mean of a column
data['Visibility_km'].mean()
```

```python
# records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km.
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Select only numeric columns for aggregation
numeric_cols = df.select_dtypes(include=['number']).columns

# Group by 'Weather_Condition' and calculate the mean for numeric columns
mean_values = df.groupby('Weather_Condition')[numeric_cols].mean()

mean_values.head()
```

```python
#instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.
df[(df['Weather_Condition'] == 'Clear') & (df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40)]
```

```python
#Show snow weather conditions
df[df['Weather_Condition'] == 'Snow']
```
